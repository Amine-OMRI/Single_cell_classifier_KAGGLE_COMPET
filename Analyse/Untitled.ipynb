```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
def plot_cell_segments(cell_mask, nuclei_mask, mt, er, nu, gr):
    
    i = 0
    microtubule = plt.imread(mt[i])    
    endoplasmicrec = plt.imread(er[i])    
    nuclei = plt.imread(nu[i])
    green=plt.imread(gr[i])
    #img = np.dstack((microtubule, endoplasmicrec, nuclei))
    
    #img = np.dstack((green,nuclei))
    
    plt.figure(figsize=(15, 15))
    plt.subplot(1, 4, 1)
    plt.imshow(green)
    plt.title('Image')
    plt.axis('off')
    
    plt.subplot(1, 4, 2)
    plt.imshow(cell_mask)
    plt.title('Cell Mask ')
    plt.axis('off')
    
    plt.subplot(1, 4, 3)
    plt.imshow(nuclei_mask)
    plt.title(' Nuclei Mask')
    plt.axis('off')
    
    plt.subplot(1, 4, 4)
    plt.imshow(green)
    plt.imshow(cell_mask, alpha=0.6)
    plt.imshow(nuclei_mask, alpha=0.6)
    plt.title('Image + Mask')
    plt.axis('off')
    plt.show()
```

```python
TRAIN_SET=os.path.join("static", "images")
def build_image_names(image_id: str, folder:str= TRAIN_SET) -> list:
    # mt is the mitchondria
    mt = os.path.join(folder, image_id + '_red.png')
    
    # er is the endoplasmic reticulum
    er = os.path.join(folder, image_id + '_yellow.png')
    
    # nu is the nuclei
    nu = os.path.join(folder, image_id + '_blue.png')
    
    # nu is the nuclei
    gr = os.path.join(folder, image_id + '_green.png')
    
    return [mt], [er], [nu] ,[gr], [[mt], [er], [nu]]
```

```python
image_id='user_image'
cell_mask_name='cell_mask'
nucl_mask_name='nucl_mask'
DIR_MASK= os.path.join("static", "images","dir_mask")

mt, er, nu, gr, images = build_image_names(image_id)    
cell_mask = np.load(f'{DIR_MASK}/{cell_mask_name}.npz')['arr_0']
nucl_mask = np.load(f'{DIR_MASK}/{nucl_mask_name}.npz')['arr_0']
plot_cell_segments(cell_mask, nucl_mask, mt, er, nu, gr)
```
